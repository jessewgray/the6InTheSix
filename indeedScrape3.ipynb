```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
number = 0
objList = []
for i in range(1,31):
    if i < 12:
        jobVar = 'analyst'
    elif i < 22:
        jobVar = 'engineer'
    else:
        jobVar = 'scientist'
    url = 'https://www.indeed.ca/jobs?q=data+' + jobVar + '&l=canada&start=' + str(number)
    response = requests.get(url)
    soup = bs(response.text, 'lxml')
    #print(number, i)
    number = number + 20
    
    resultList = []
    results=soup.find_all('div', class_='jobsearch-SerpJobCard')
    
    #objList = []

    for j in range(len(results)):
        jobTitle = results[j].a.text.lower()
        company = results[j].span.text.strip()
        if results[j].find('div', class_="location"):
            location = results[j].find('div', class_="location")
            location = location.text
        elif results[j].find('span', class_="location"):
            location = results[j].find('span', class_="location")
            location = location.text
        splitString = location.split(',')
        if len(splitString) < 2:
            city = splitString[0].strip()
            province = splitString[0].strip()
        else:
            city = splitString[0].strip()
            province = splitString[1].strip()
        if results[j].find('span', class_="salary"):
            salary = results[j].find('span', class_="salary")
            salary = salary.text.strip()
        else:
            salary = "n/a"
        summary = results[j].find('span', class_="summary")
        summary = summary.text.strip()
        theLink = results[0].a['href']
        theLink = "https://www.indeed.ca" + theLink
        if "analyst" in jobTitle:
            jobID = 1
        elif "engineer" in jobTitle:
            jobID = 2
        elif "scientist" in jobTitle:
            jobID = 3
        else:
            jobID = "n/a"
        obj = {
            "jobTitle": jobTitle,
            "jobID": jobID,
            "company": company,
            "city": city, 
            "province": province,
            "salary": salary,
            "summary":summary,
            "jobLink":theLink
        }
        objList.append(obj)
        resultList = []
        if number == 200:
            number = 0
#objList
```

```python
indeedDF = pd.DataFrame(objList)
indeedDF
```

```python
indeedDF.to_csv("./1_Extracted_Inputs/indeedCSV.csv", index = False)
```
