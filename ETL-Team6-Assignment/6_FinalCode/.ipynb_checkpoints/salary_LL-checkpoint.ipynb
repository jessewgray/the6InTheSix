```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
jo_id = []
jo_tit = []
jo_sal = []
jo_sal_desc = []
jo_prov = []
```

```python
count = 0

titles_list = ['Data-Analyst',
          'Data-Engineer',
          'Data-Scientist']

prov = ['Ontario',
        'Alberta',
        'British-Columbia',
        'Québec-Province',
        'Saskatchewan',
        'Yukon',
        'Nova-Scotia',
        'Manitoba',
        'New-Brunswick',
        'Newfoundland-And-Labrador',
        'Prince-Edward-Island',
        'Northwest-Territories',
        'Nunavut']

for idx,val in enumerate(titles_list):
    pages = True
    for province in prov:
        while pages == True:
            url = f"https://www.indeed.ca/salaries/{val}-Salaries,-{province}?start={count}"
            response = requests.get(url)
            soup = bs(response.text, 'lxml')

            if soup.find('table', class_='cmp-TitleSalary-table'):
                table = soup.find('table', class_='cmp-TitleSalary-table')
                titles = table.find_all('span', class_= None)
                salaries = table.find_all('span', class_= 'cmp-salary-amount')
                salary_desc = table.find_all('div', class_= 'cmp-sal-summary')
                
                jo_id.extend([idx + 1] * (len(titles)))
                jo_prov.extend([province] * (len(titles)))
                jo_tit.extend(title.text for title in titles)
                jo_sal.extend(float(salary.text[1:].replace(',','')) for salary in salaries)
                jo_sal_desc.extend(desc.text[-8:] for desc in salary_desc)
                
                if soup.find('span', class_='cmp-Pagination-link'):
                    count += 10
                    continue
                else:
                    break
            else:
                count = 0
                break
```

```python
df = pd.DataFrame(data={'ID':jo_id,
                        'Title': jo_tit,
                        'Province': jo_prov,
                        'Annual_salary': jo_sal,
                        'typ':jo_sal_desc})

df
```

```python
df['Annual_salary'].loc[df['typ'].str.contains('month')] = df['Annual_salary'] * 12
df['Annual_salary'].loc[df['typ'].str.contains('week')] = df['Annual_salary'] * 52
df['Annual_salary'].loc[df['typ'].str.contains('hour')] = df['Annual_salary'] * 2080
df
```

```python
df_fin = df.drop(['typ'], axis = 1)

df_fin.to_csv(r'../1_Extracted_Inputs/Salary_extracted_input.csv', index = False)
```
